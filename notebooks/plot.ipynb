```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
names = ["SpyNet",	"MSE Pyr",	"Metr Pyr",	"MSE Pyr+opt",	"Metr Pyr+opt"]
mean_test = np.array([0.431,	0.197,	0.135,	0.111,	0.104])
mean_train = np.array([0.47,	0.23,	0.16,	0.12,	0.10])
stdev_train = np.array([0.15,	0.03,	0.02,	0.02,	0.02])
stdev_test = np.array([0.182,	0.018,	0.012,	0.011,	0.012])
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects = names
y_pos = np.arange(len(objects))

plt.figure(figsize=(7,5))

plt.bar(y_pos, mean_train, yerr=stdev_train, 
        align='center', alpha=0.4, ecolor='black', capsize=10,
       width=0.5)

plt.xticks(y_pos, objects)
plt.ylabel('Fraction of Unaligned Cable Lengths')
plt.title('Train Set Cable Length Fraction')
plt.tick_params(axis='x', which='major', labelsize=10, pad=15, width=10)
#plt.figure(figsize=(15,5))
#plt.xticks(rotation=-15)
plt.show()
```

```python


objects = names
y_pos = np.arange(len(objects))

plt.figure(figsize=(7,5))

plt.bar(y_pos, mean_test, yerr=stdev_test, 
        align='center', alpha=0.4, ecolor='black', capsize=10,
       width=0.5, color='salmon')

plt.xticks(y_pos, objects)
plt.ylabel('Fraction of Unaligned Cable Lengths')
plt.title('Test Set Cable Length Fraction')
plt.tick_params(axis='x', which='major', labelsize=10, pad=15, width=10)
#plt.figure(figsize=(15,5))
#plt.xticks(rotation=-15) ny7
plt.show()
```

```python


objects = names
y_pos = np.arange(len(objects))

plt.figure(figsize=(9, 7))

plt.bar(y_pos-0.3, mean_train, yerr=stdev_train, 
        align='center', alpha=0.4, ecolor='black', capsize=10,
       width=0.3)

plt.bar(y_pos, mean_test, yerr=stdev_test, 
        align='center', alpha=0.4, ecolor='black', capsize=10,
       width=0.3, color='salmon', label='Test Set')

plt.xticks(y_pos, objects)
plt.ylabel('Fraction of Unaligned Cable Lengths')
plt.title('Test and Train Set Cable Length Fraction')
plt.tick_params(axis='x', which='major', labelsize=10, pad=15, width=10)
#plt.figure(figsize=(15,5))
#plt.xticks(rotation=-15) ny7
plt.show()
```

```python


objects = names
y_pos = np.arange(len(objects[3:]))

plt.figure(figsize=(9, 7))

plt.bar(y_pos-0.2, mean_train[3:]/mean_train.min(), 
        align='center', alpha=0.4, ecolor='black',
       width=0.2)

plt.bar(y_pos, mean_test[3:]/mean_test.min(), 
        align='center', alpha=0.4, ecolor='black',
       width=0.2, color='salmon', label='Test Set')

plt.xticks(y_pos, objects)
plt.ylabel('Fraction of Best Cable Lengths')
plt.title('Test and Train Set Cable Length Fraction')
plt.tick_params(axis='x', which='major', labelsize=10, pad=15, width=10)
#plt.figure(figsize=(15,5))
#plt.xticks(rotation=-15) ny7
plt.show()
```
